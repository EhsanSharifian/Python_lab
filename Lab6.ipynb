```python
class Set():
    def __init__(self):
        self.list = []

    def add(self, x):
        if x not in self.list:
            self.list.append(x)

    def delete(self, x):
        if x in self.list:
            self.list.remove(x)

    def __or__(self, other):
        union = Set()
        for member in self.list:
            union.add(member)
        for member in other.list:
            union.add(member)
        return union

    def __and__(self, other):
        intersect = Set()
        for member in self.list:
            if member in other.list:
                intersect.add(member)
        return intersect

    def __eq__(self, other):
        return set(self.list) == set(other.list)


a = Set()
a.add('Faraz')
a.add('Ehsan')
b = Set()
b.add('Pouria')
b.add('Rouzbeh')
c = a | b
print("set a = ", a.list)
print("set b = ", b.list)
print("c = a|b = ", c.list)
d = c & a
print("d = a&c = ", d.list)
print("Are a & b equal?",a == b)
print("Are a & d equal?",a == d)
c.delete('Faraz')
print("c without Faraz",c.list)
```
